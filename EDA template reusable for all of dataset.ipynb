# 🚀 EDA TEMPLATE (REUSABLE)

1. 🔧 Setup & Load Data

```python
# ======================
# 1. SETUP
# ======================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="pastel")
plt.rcParams["figure.figsize"] = (8,4)

# ======================
# 2. LOAD DATA
# ======================
def load_data(path):
    if path.endswith(".csv"):
        return pd.read_csv(path)
    elif path.endswith(".xlsx"):
        return pd.read_excel(path)
    else:
        raise ValueError("Unsupported file format")

df = load_data("your_data.csv")  # 👈 CHANGE HERE
```

2. 📊 Basic Overview (ALWAYS FIRST)

```python
def basic_overview(df):
    print("Shape:", df.shape)
    print("\nData Types:\n", df.dtypes)
    print("\nMissing Values:\n", df.isnull().sum().sort_values(ascending=False))
    print("\nDuplicate Rows:", df.duplicated().sum())

basic_overview(df)
```

👉 Insight:

Có đáng tin không (Data Quality – nhóm X)

3. 📈 Summary Statistics

```python
def summary_stats(df):
    return df.describe().T

summary_stats(df)
```

👉 Nhìn:
- Mean vs Median → skew
- Std → variance

4. 🔍 Auto Detect Column Types

```python
def detect_columns(df):
    num_cols = df.select_dtypes(include=np.number).columns.tolist()
    cat_cols = df.select_dtypes(exclude=np.number).columns.tolist()
    return num_cols, cat_cols

num_cols, cat_cols = detect_columns(df)
```

5. 📊 Distribution (Numerical)

```python
def plot_numeric_distribution(df, cols):
    for col in cols:
        plt.figure()
        sns.histplot(df[col].dropna(), kde=True)
        plt.title(f"Distribution of {col}")
        plt.show()

plot_numeric_distribution(df, num_cols)
```

👉 Insight:
- Outliers
- Long tail
- Skew

6. 📦 Boxplot (Outlier detection)

```python
def plot_boxplots(df, cols):
    for col in cols:
        plt.figure()
        sns.boxplot(x=df[col])
        plt.title(f"Boxplot of {col}")
        plt.show()

plot_boxplots(df, num_cols)
```

7. 🧩 Categorical Analysis

```python
def plot_categorical(df, cols):
    for col in cols:
        plt.figure()
        sns.countplot(x=col, data=df, order=df[col].value_counts().index)
        plt.xticks(rotation=45)
        plt.title(f"Count of {col}")
        plt.show()

plot_categorical(df, cat_cols)
```

👉 Insight:
- Imbalance
- Dominant category

8. 🔥 Correlation Matrix

```python
def plot_correlation(df):
    corr = df.corr(numeric_only=True)
    sns.heatmap(corr, annot=True, cmap="coolwarm")
    plt.title("Correlation Matrix")
    plt.show()

plot_correlation(df)
```

9. 🔗 Relationship (Scatter)

```python
def plot_relationship(df, num_cols):
    for i in range(len(num_cols)):
        for j in range(i+1, len(num_cols)):
            plt.figure()
            sns.scatterplot(data=df, x=num_cols[i], y=num_cols[j])
            plt.title(f"{num_cols[i]} vs {num_cols[j]}")
            plt.show()

plot_relationship(df, num_cols)
```

10. 🎯 Target Analysis (QUAN TRỌNG)

👉 Bạn chỉ cần set target:

```python
TARGET = "converted"  # 👈 CHANGE HERE
```

Numerical vs Target

```python
def target_vs_numeric(df, num_cols, target):
    for col in num_cols:
        if col != target:
            plt.figure()
            sns.boxplot(x=target, y=col, data=df)
            plt.title(f"{col} vs {target}")
            plt.show()

target_vs_numeric(df, num_cols, TARGET)
```

Categorical vs Target

```python
def target_vs_category(df, cat_cols, target):
    for col in cat_cols:
        plt.figure()
        sns.barplot(x=col, y=target, data=df)
        plt.xticks(rotation=45)
        plt.title(f"{target} by {col}")
        plt.show()

target_vs_category(df, cat_cols, TARGET)
```

👉 Insight:

Feature nào ảnh hưởng mạnh nhất

11. 🕒 Time Analysis (OPTIONAL)

```python
def time_analysis(df, date_col, value_col):
    df[date_col] = pd.to_datetime(df[date_col])
    temp = df.groupby(date_col)[value_col].sum().reset_index()

    sns.lineplot(data=temp, x=date_col, y=value_col)
    plt.title("Time Trend")
    plt.show()

# time_analysis(df, "date", "revenue")  # 👈 USE IF HAVE DATE
```

12. 🧠 Quick Insights Generator (AUTO)

```python
def quick_insights(df):
    print("\n=== QUICK INSIGHTS ===")

    # Skew check
    for col in df.select_dtypes(include=np.number).columns:
        skew = df[col].skew()
        if abs(skew) > 1:
            print(f"{col} is highly skewed ({skew:.2f})")

    # Missing warning
    missing = df.isnull().sum()
    for col in missing.index:
        if missing[col] > 0:
            print(f"{col} has missing values")

quick_insights(df)
```
